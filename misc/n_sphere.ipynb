```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Define the function to calculate the volume of an n-dimensional sphere
def sphere_volume(n):
    return np.pi**(n / 2) / gamma(n / 2 + 1)

# Generate values for n (number of dimensions)
n_values = np.arange(1, 25)  # Dimensions from 1 to 20
volumes = sphere_volume(n_values)

# Plot the volume as a function of dimensions
plt.figure(figsize=(8, 6))
plt.plot(n_values, volumes, marker='o', linestyle='-', color='b', label="Volume of n-D Sphere")
plt.title("Volume of an n-Dimensional Sphere (Unit Radius)", fontsize=14)
plt.xlabel("Number of Dimensions (n)", fontsize=12)
plt.ylabel("Volume", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
